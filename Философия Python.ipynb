### 3.2. Duck Typing ("–ï—Å–ª–∏ —ç—Ç–æ —Ö–æ–¥–∏—Ç –∫–∞–∫ —É—Ç–∫–∞ –∏ –∫—Ä—è–∫–∞–µ—Ç –∫–∞–∫ —É—Ç–∫–∞...")

–ù–µ –≤–∞–∂–µ–Ω —Ç–∏–ø –æ–±—ä–µ–∫—Ç–∞, –≤–∞–∂–Ω—ã –µ–≥–æ –º–µ—Ç–æ–¥—ã –∏ –∞—Ç—Ä–∏–±—É—Ç—ã.

````python
class Duck:
    def quack(self):
        return "Quack!"

class Person:
    def quack(self):
        return "I'm quacking like a duck!"

def make_it_quack(thing):
    # –ù–∞—Å –Ω–µ –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç —Ç–∏–ø, —Ç–æ–ª—å–∫–æ –Ω–∞–ª–∏—á–∏–µ –º–µ—Ç–æ–¥–∞ quack()
    print(thing.quack())

make_it_quack(Duck())    # –†–∞–±–æ—Ç–∞–µ—Ç
make_it_quack(Person())  # –¢–æ–∂–µ —Ä–∞–±–æ—Ç–∞–µ—Ç!
````

### 3.3. EAFP: Easier to Ask for Forgiveness than Permission

–ü—Ä–æ—â–µ –ø—Ä–æ—Å–∏—Ç—å –ø—Ä–æ—â–µ–Ω–∏—è, —á–µ–º —Ä–∞–∑—Ä–µ—à–µ–Ω–∏—è.

````python
my_dict = {"name": "Python", "version": "3.11"}

# EAFP —Å—Ç–∏–ª—å (Python way)
try:
    value = my_dict['key']
except KeyError:
    value = 'default'
    print(f"EAFP: {value}")

# LBYL —Å—Ç–∏–ª—å (Look Before You Leap)
if 'key' in my_dict:
    value = my_dict['key']
else:
    value = 'default'
    print(f"LBYL: {value}")
````

### –ü—Ä–∏–º–µ—Ä 1: –ß–∏—Ç–∞–µ–º–æ—Å—Ç—å –∫–æ–¥–∞

**–ó–∞–¥–∞—á–∞:** –ù–∞–π—Ç–∏ –≤—Å–µ —á—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –∏ –≤–æ–∑–≤–µ—Å—Ç–∏ –∏—Ö –≤ –∫–≤–∞–¥—Ä–∞—Ç.

````python
# –ü–ª–æ—Ö–æ (–Ω–µ —Å–ª–µ–¥—É–µ—Ç —Ñ–∏–ª–æ—Å–æ—Ñ–∏–∏ Python)
def f(l):
    r=[]
    for i in l:
        if i%2==0:
            r.append(i*i)
    return r

print(f"–ü–ª–æ—Ö–æ: {f([1, 2, 3, 4, 5, 6])}")

# –•–æ—Ä–æ—à–æ (—Å–ª–µ–¥—É–µ—Ç —Ñ–∏–ª–æ—Å–æ—Ñ–∏–∏ Python)
def get_squares_of_even_numbers(numbers):
    return [number ** 2 for number in numbers if number % 2 == 0]

print(f"–•–æ—Ä–æ—à–æ: {get_squares_of_even_numbers([1, 2, 3, 4, 5, 6])}")
````

### –ü—Ä–∏–º–µ—Ä 3: –ü—Ä–æ—Å—Ç–æ—Ç–∞

````python
# –°–ª–æ–∂–Ω–æ
def get_max_complex(numbers):
    if len(numbers) == 0:
        return None
    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = num
    return max_value

# –ü—Ä–æ—Å—Ç–æ
def get_max(numbers):
    return max(numbers) if numbers else None

test_numbers = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"–°–ª–æ–∂–Ω–æ: {get_max_complex(test_numbers)}")
print(f"–ü—Ä–æ—Å—Ç–æ: {get_max(test_numbers)}")
````

### –ü—Ä–∏–º–µ—Ä 4: –û–±—Ä–∞–±–æ—Ç–∫–∞ –æ—à–∏–±–æ–∫

````python
# –ü–ª–æ—Ö–æ - –æ—à–∏–±–∫–∞ –∏–≥–Ω–æ—Ä–∏—Ä—É–µ—Ç—Å—è
def divide_bad(a, b):
    try:
        return a / b
    except:
        return None

# –•–æ—Ä–æ—à–æ - –æ—à–∏–±–∫–∞ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç—Å—è —è–≤–Ω–æ
def divide_good(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        raise ValueError("Cannot divide by zero")

print(f"divide_bad(10, 0): {divide_bad(10, 0)}")

try:
    divide_good(10, 0)
except ValueError as e:
    print(f"divide_good(10, 0): {e}")
````
