### ๐ ะัะธะผะตัั ัะตััะพะฒ ะดะปั ัะฐะทะฝัั ัะธััะฐัะธะน

#### 1๏ธโฃ ะขะตััะธัะพะฒะฐะฝะธะต ะฟัะพััะพะน ััะฝะบัะธะธ

````python
# ะขะตััะธััะตะผะฐั ััะฝะบัะธั
def calculate_discount(price: float, discount_percent: float) -> float:
    """ะะฐัััะธัะฐัั ัะตะฝั ัะพ ัะบะธะดะบะพะน."""
    if discount_percent < 0 or discount_percent > 100:
        raise ValueError("ะกะบะธะดะบะฐ ะดะพะปะถะฝะฐ ะฑััั ะพั 0 ะดะพ 100%")
    return price * (1 - discount_percent / 100)

# ะขะตััั
def test_discount_10_percent():
    """ะกะบะธะดะบะฐ 10% ะพั 1000 = 900."""
    result = calculate_discount(1000, 10)
    assert result == 900

def test_discount_zero():
    """ะะตะท ัะบะธะดะบะธ ัะตะฝะฐ ะฝะต ะผะตะฝัะตััั."""
    result = calculate_discount(500, 0)
    assert result == 500

def test_discount_100_percent():
    """ะกะบะธะดะบะฐ 100% = ะฑะตัะฟะปะฐัะฝะพ."""
    result = calculate_discount(1000, 100)
    assert result == 0

def test_discount_negative_raises_error():
    """ะััะธัะฐัะตะปัะฝะฐั ัะบะธะดะบะฐ โ ะพัะธะฑะบะฐ."""
    import pytest
    with pytest.raises(ValueError):
        calculate_discount(1000, -10)
````

#### 2๏ธโฃ ะขะตััะธัะพะฒะฐะฝะธะต ัััะพะบะพะฒัั ััะฝะบัะธะน

````python
import re

# ะขะตััะธััะตะผะฐั ััะฝะบัะธั
def slugify(text: str) -> str:
    """ะัะตะพะฑัะฐะทะพะฒะฐัั ัะตะบัั ะฒ URL-slug."""
    text = text.lower().strip()
    text = re.sub(r'[^\w\s-]', '', text)
    text = re.sub(r'[\s_-]+', '-', text)
    return text

# ะขะตััั
def test_slugify_simple():
    assert slugify("Hello World") == "hello-world"

def test_slugify_with_special_chars():
    assert slugify("Hello, World!") == "hello-world"

def test_slugify_with_spaces():
    assert slugify("  Multiple   Spaces  ") == "multiple-spaces"

def test_slugify_already_slug():
    assert slugify("already-a-slug") == "already-a-slug"

def test_slugify_empty_string():
    assert slugify("") == ""

def test_slugify_russian():
    assert slugify("ะัะธะฒะตั ะะธั") == "ะฟัะธะฒะตั-ะผะธั"
````
