## 14.1 Файловый ввод-вывод

**Файл** - блок данных, который хранится на внешнем носителе (жесткий диск, флеш-накопитель) как единое целое. Является абстракцией операционной системы (ОС), скрывающей от программ (и людей) особенности хранилища. Т.е. взаимодействие с файлом (вне зависимости от того как он представлен непосредственно на носителе) осуществляется через стандартизированный интерфейс операционной системы.
Свойства файла и различные ограничения, связанные с файлом, во многом определяются **файловой системой**.
<br/>
<br/>

**Файловая система** - набор правил, механизмов и структур данных, определяющих способ организации, хранения, именования и доступа к данным на носителе.
В большинстве файловых систем определены: *имя файла*, *тип* файла (расширение) и *путь* до файла.

*Имя файла* используется для уникальной идентификации (т.е. выполняет те же задачи, что и имя переменной в коде) файла.

*Тип файла* позволяет указать на ожидаемый способ обработки файла и его внутреннюю структуру (так, расширение .docx говорит о том, что это файл текстового редактора MS Word). Не является обязательным элементом имени, т.е. возможно существование файла и без расширения.

*Путь к файлу* позволяет однозначно определить расположение файла в иерархии **каталогов** файловой системы.
<br/>
<br/>

**Каталог** (директория или папка) - специальный элемент (по сути, ещё один файл) файловой системы, упрощающий организацию данных. Позволяет группировать файлы и выстраивать иерархии.
<br/>
<br/>
Если мы говорим о некоторой иерархии файлов, нельзя не сказать про *абсолютный* и *относительный* пути к файлу. Разберем эти понятия на примере. Допустим, мы имеем следующую иерархию каталогов и файлов:

<pre>
C:
|
-Code
  |
  -test.py
  -Python
    |
    -Тема 1
      |
      -Task1.py
      -Task2.py
    -Тема 2
      |
      -Task1.py
      -Task2.py
</pre>

Если мы открыли каталог "Тема 1", то в рамках этого каталога *относительный* путь к файлу "Task1.py" будет записан как:

<code>./Task1.py</code>

, где **'./'** - указывает на то, что искать файл необходимо в текущем каталоге.

*Абсолютный* путь к файлу "Task1.py" будет записан как:

<code>C:/Code/Python/Тема 1/Task1.py</code>

"Находясь" в каталоге "Тема 1" попробуем построить *относительный* путь к файлу "Task2.py" из каталога "Тема 2":

<code>../Тема 2/Task2.py</code>

, где **'../'** - указывает о необходимости начать поиск не в текущем каталоге, а в родительском (по отношению к текущему) каталоге. Родительским каталогом для "Тема 1" будет "Python".

Попробуем прописать относительный путь из "Тема 1" к файлу "test.py":

<code>../../test.py</code>

Важно, что *относительный* путь всегда строится из некоторой точки, поэтому при изменении начальной точки, нужно менять и относительный путь. Так, находясь в каталоге "С:" обратиться к файлу "Test.py" можно через относительный путь:

<code>./Code/test.py</code>

Т.е. файл "test.py" не менял своего расположения, но относительный путь к нему изменился из-за смены места "старта".

Абсолютный путь к файлу "test.py" всегда будет одинаковым:

<code>C:/Code/test.py</code>
<br/>

При работе с файлами через программный код, важно уметь простраивать относительные пути и избегать использования абсолютных путей.

<br/>

### 14.1.1 Работа с файлами

Python позволяет работать с файлами. Общий алгоритм обработки какого-либо файла в Python выглядит так:
1. Создать переменную.
2. Связать её с открытым файлом с помощью функции <code>open</code>.
3. Воспользоваться набором базовых операций (.write(), .read()).
4. **Закрыть файл**.

Разберем на примере:

```python
# Простая работа с файлом

# Связываем переменную с файлом simple_file.txt
# При использовании относительных путей, нужно учитывать, что "стартовым" каталогом будет каталог, где расположен запущенный файл .py
file = open("./simple_file.txt", encoding="utf-8")

# Выполняем операцию чтения
file_data = file.read()
print(f"В файле найден текст: '{file_data}'")

# Закрываем файл
file.close()
```

```python
# Попробуем выполнить запись в файл

# Связываем переменную с файлом simple_file.txt
file = open("./simple_file.txt", encoding="utf-8")

# Выполняем операцию записи
file.write("Новые данные\n")

# Закрываем файл
file.close()
```

В коде выше мы получили ошибку при выполнении записи в файл. Дело в том, что функция open ожидает не только название файла, но и режим открытия. Если мы не указываем режим открытия файла, файл открывается в режиме "r" (сокращение от "read", т.е. чтение).

Вот какие режимы открытия файла есть:
<ul>
<li>"r" - чтение (применяется по умолчанию)</li>
<li>"w" - запись (создает новый файл, если его не существует; очищает файл, если файл существовал до вызова open)</li>
<li>"a" - дозапись в конец</li>
<li>"x" - запись (создает новый файл, если его не существует; выдаёт ошибку, если файл с таким именем уже есть)</li>
</ul>
Модификаторы режимов (могут быть использованы вместе с одним из базовых режимов):
<ul>
<li>"+" - обновление (позволяет и читать, и изменять файл)</li>
<li>"b" - работа с бинарным файлом (функции чтения и записи будут ожидать объект bytes, а не str)</li>
<li>"t" - работа с текстовым файлом (применяется по умолчанию, после открытия Python автоматически применяет кодировку, чтобы прочитать не "сырые" биты, а символы текста)</li>
</ul>

Рассмотрим примеры работы с файлом.

```python
# Снова пробуем выполнить запись в файл

# Связываем переменную с файлом simple_file.txt
# Режим: запись с обязательным удалением данных в файле
file = open("./simple_file.txt", mode="w", encoding="utf-8")

# Выполняем операцию записи
file.write("Новые данные\n")

# Закрываем файл
file.close()

# Прочитаем то, что записано
file = open("./simple_file.txt", mode="r", encoding="utf-8")
file_data = file.read()
print(f"Текущее содержимое файла:\n'{file_data}'")
file.close()
```

```python
# Не очень хорошо, что файл очищается перед записью. Попробуем исправить

# Связываем переменную с файлом simple_file.txt
# Режим: дозапись в конец
file = open("./simple_file.txt", mode="a", encoding="utf-8")

# Выполняем операцию записи
file.write("Новые данные\n")
# file.write(10) # Ошибка: в текстовый файл можно записывать только строковые значения

# Закрываем файл
file.close()

# Прочитаем то, что записано
file = open("./simple_file.txt", mode="r", encoding="utf-8")
file_data = file.read()
print(f"Текущее содержимое файла:\n'{file_data}'")
file.close()
```

```python
# Для записи и прочтения данных из файла нам приходится дважды открывать и закрывать файл. Это не очень удобно. Попробуем исправить

# Связываем переменную с файлом simple_file.txt
# Режим: чтение с возможностью обновления
file = open("./simple_file.txt", mode="r+", encoding="utf-8")

# Выполняем операцию записи
file.write("Самые новые данные\n")

file_data = file.read()
print(f"Содержимое файла:\n'{file_data}'")

# Закрываем файл
file.close()
```

В последнем примере получилось что-то странное. При выводе содержимого файла мы получили только кусочек текста. Больше того, если открыть файл, можно увидеть, что новая запись (текст "Самые новые данные\n") была выполнена "поверх" старой строки "Новые данные". Почему?

Ответ: операции записи и чтения всегда выполняются из позиции *курсора*.

Курсор это значение *текущего смещения* относительно начала файла (начало файла = смещение 0). Когда мы выполняем операции записи и чтения, курсор сдвигается (всегда вправо, т.е. в сторону конца файла). Важно учитывать, что **смещение всегда указывается в байтах**, а не в символах.

Рассмотрим перемещения курсора из кусочка кода выше:
1. Файл открыт в режиме чтения с обновлением, курсор установлен в позицию 0 (начало).
2. Выполняется запись (.write("Самые новые данные\n")). Курсор смещается вправо на количество байт, которое требуется для записываемой строки.
3. Выполняется чтение (.read). Курсор смещается в конец файла.

Отсюда можно сделать два вывода\предположения: 
1. Если после записи мы сможем сдвинуть курсор обратно в позицию 0, при последующем считывании мы получим полное содержимое файла.
2. Если после прочтения всего файла снова попытаться вызвать операцию чтения, читать будет нечего, так как курсор после первого вызова .read() будет смещен в конец файла.

Проверим наши предположения:

```python
# Используем метод .seek(смещение, старт_смещения) для смещения курсора в начало файла

# Связываем переменную с файлом simple_file.txt
# Режим: чтение с возможностью обновления
file = open("./simple_file.txt", mode="r+", encoding="utf-8")

# Выполняем операцию записи
file.write("Самые новые данные\n")

# Выполняем смещение в позицию 0
file.seek(0)

file_data = file.read()
print(f"Содержимое файла:\n'{file_data}'")

# Закрываем файл
file.close()
```

```python
# Пытаемся дважды прочитать файл

# Связываем переменную с файлом simple_file.txt
# Режим: чтение
file = open("./simple_file.txt", mode="r", encoding="utf-8")

# Первое чтение
file_data = file.read()
print(f"Содержимое файла (1 вызов .read()):\n'{file_data}'\n")

# Читаем повторно
new_file_data = file.read()
print(f"Содержимое файла (2 вызов .read()):\n'{new_file_data}'")

# Закрываем файл
file.close()
```

Метод .seek() позволяет выполнять довольно тонкую настройку смещения в формате: количество бит смещения, стартовая позиция для смещения (по умолчанию принимается за 0, т.е. за начало файла).

Но нужно понимать, что все значения указывают не количество символов, а количество байт. Т.е. для правильного смещения в текстовом файле, нужно точно понимать используемую кодировку символов и размер каждого символа (разные символы могут иметь разные размеры в байтах). При работе с текстовыми файлами Python накладывает дополнительные ограничения на работу .seek(). Поэтому в большинстве случаев следует использовать только такие вызовы .seek:

1. <code>.seek(0)</code> - быстрое смещение в начало файла.
2. <code>.seek(0, 2)</code> - быстрое смещение в конец файла.

```python
# Все варианты чтения

file = open("./read_methods.txt", encoding="utf-8")

# Метод .read() - считывает все строки в одну результирующую
all_lines_str = file.read()
print(f"Метод .read():\n{all_lines_str}\n")

file.seek(0)

# Метод .readline() - считывает одну строку
one_line = file.readline()
print(f"Метод .readline():\n{one_line}\n")

file.seek(0)

# Метод .readlines() - считывает все строки по отдельности, возвращая список
all_lines_list = file.readlines()
print(f"Метод .readlines():\n{all_lines_list}")

file.close()
```

```python
# Все варианты записи
data_to_write = [
    "1. Моя строка 1\n",
    "2. Моя строка 2\n",
    "3. Моя строка 3\n",
]

file = open("./write_methods.txt", "w", encoding="utf-8")

# Метод .write(строка) - выполняет запись и ожидает, что на вход будет передано строковое значение 
file.write(data_to_write[0])
# file.write(data_to_write) # Ошибка: на вход передана не строка, а list


# Метод .writelines(коллекция с элементами str) - выполняет запись строк из некоторой коллекции
file.writelines(data_to_write)
# file.writelines([1, 2, 3, 4]) # Ошибка: коллекция содержит что-то отличное от типа str

file.close()
```

**Как только работа с файлом была завершена, его обязательно нужно закрыть с помощью метода .close()!**

Если оставить файл открытым возможны нежелательные последствия в виде:
1. Запрет от операционной системы на открытие новых файлов (если программа открыла очень много файлов и не закрыла ни один из них).
2. Утрата обновлений файла (т.е. на устройстве (на диске) останется версия файла до его открытия программой).
3. Искажение данных в файле (искажения могут затронуть как "старые" данные, так и обновленные).

К счастью, Python заботится о программистах, поэтому в его составе есть конструкция, которая позволяет избавиться от необходимости вручную закрывать файлы. Эта конструкция называется контекстным менеджером (with).

### 14.1.2 Контекстный менеджер

**Контекстный менеджер** - механизм языка Python, который позволяет гарантировать корректное управление некоторым ресурсом.

Документация: https://docs.python.org/3/library/contextlib.html

Синтаксис применения контекстного менеджера при работе с файлами:
<pre>
<code>
with open(...) as имя_переменной_для_файла:
    тело контекстного менеджера, где выполняется работа с файлом
</code>
</pre>

Можно заметить, что работа с файлом помещается в отдельный блок кода, выделенный отступами. Этот блок и определяет то "время", где файл будет открыт. Как только исполняемый поток выйдет из этого блока (в случае выполнения всех строк в теле контекстного менеджера или в случае возникновения ошибки), контекстный менеджер выполнит операции для корректного закрытия файла.

Посмотрим на примере (для наглядности возьмем код предыдущего блока):

```python
# Все варианты записи (версия с контекстным менеджером)

data_to_write = [
    "1. Моя строка 1\n",
    "2. Моя строка 2\n",
    "3. Моя строка 3\n",
]

with open("./write_methods.txt", "w", encoding="utf-8") as file:
    # Метод .write(строка) - выполняет запись и ожидает, что на вход будет передано строковое значение 
    file.write(data_to_write[0])
    # file.write(data_to_write) # Ошибка: на вход передана не строка, а list


    # Метод .writelines(коллекция с элементами str) - выполняет запись строк из некоторой коллекции
    file.writelines(data_to_write[1:])
    # file.writelines([1, 2, 3, 4]) # Ошибка: коллекция содержит что-то отличное от типа str

print('All set! File is closed')
```

```python
# Контекстный менеджер и открытие нескольких файлов

with (
    open("./read_methods.txt", encoding="utf-8") as file_1,
    open("./simple_file.txt", encoding="utf-8") as file_2,
    open("./write_methods.txt", encoding="utf-8") as file_3,
    ):
    print(f"Содержимое файла 1:\n{file_1.read()}")
    print(f"Содержимое файла 2:\n{file_2.read()}")
    print(f"Содержимое файла 3:\n{file_3.read()}")

print('All set! All files are closed')
```

```python
# Использование контекстного менеджера для работы с контекстом модуля decimal

from decimal import Decimal, localcontext

print(Decimal("1") / Decimal("28"))

with localcontext() as context:
    # Задаем точность вычислений
    context.prec = 56
    print(Decimal("1") / Decimal("28"))

print(Decimal("1") / Decimal("28"))
```

Для работы с with можно создавать пользовательские классы. Для этого в классе необходимо реализовать "магические" методы <code>enter</code> и <code>exit</code>.

```python
import time

class Timer:
    def __init__(self):
        self.start = 0
        self.end = 0
        self.total = 0

    def __enter__(self):
        self.start = time.perf_counter()
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.stop = time.perf_counter()
        self.total = self.stop - self.start
```

```python
timer = Timer()
with timer:
    result = sum(range(100_500_000))

print(f'Total time spent: {timer.total:.2f} sec')
```

```python
with Timer() as timer:
    print(f'Start at {timer.start}')
    result = sum(range(100_500_000))

print(f'Total time spent: {timer.total:.2f} sec')
```

Пример использования класса для работы с переменными окружения в рамках контекста:

```python
import os

class set_env_var:
    def __init__(self, var_name, new_value):
        self.var_name = var_name
        self.new_value = new_value

    def __enter__(self):
        self.original_value = os.environ.get(self.var_name)
        os.environ[self.var_name] = self.new_value

    def __exit__(self, exc_type, exc_val, exc_tb):
        if self.original_value is None:
            del os.environ[self.var_name]
        else:
            os.environ[self.var_name] = self.original_value
```

```python
os.environ['active_user'] = 'Petr'
```

```python
print(f'Active user: {os.environ["active_user"]}')
with set_env_var('active_user', 'Ivan'):
    print(f'Active context user: {os.environ["active_user"]}')

print(f'Active context user after context: {os.environ["active_user"]}')
```

Контекстный менеджер можно реализовать через генераторную функцию с помощью декоратора <code>contextmanager</code>:

```python
from contextlib import contextmanager

@contextmanager
def writable_file(file_path):
    file = open(file_path, mode="w", encoding="utf-8")
    try:
        yield file
    finally:
        file.close()


users = ['Ivan', 'Petr']
with writable_file('users.txt') as file:
    file.write('\n'.join(users))
```

```python
from contextlib import contextmanager

@contextmanager
def hello_context_manager():
     print("Entering the context...")
     yield 12
     print("Leaving the context...")


with hello_context_manager() as hello:
     print(hello)
```

## 14.2 Основные стандартные модули

### 14.2.1 csv

Рассмотрим работу с простыми табличными файлами формата CSV (Comma Separated Values - Значения Разделенные Запятыми).

CSV - это довольно популярный формат, когда речь идёт о необходимости передать и сохранить большой объем структурированных данных. Формат CSV предполагает хранение данных в виде таблицы:

1. Первая строка содержит заголовок (т.е. названия столбцов)
2. Все последующие строки задержат значения для каждого столбца

Значения столбцов отделены друг от друга *разделителем*, т.е. специальным символом. По умолчанию это символ запятой. Вот так может выглядеть файл CSV внутри:

<pre>
Имя,Фамилия,Группа
Иван,Иванов,434-1
Петр,Петров,643-2
</pre>

Если открыть такой файл в табличном редакторе (MS Excel или LibreOffice Calc), мы увидим привычные таблички, где можем использовать формулы для вычисления значений в столбцах и добавлять графики (*только для сохранения этих графиков придётся выполнить преобразование файла в формат .xlsx*).

Работать в коде с файлами CSV можно и без дополнительных модулей, но это не слишком удобно. Поэтому в стандартной библиотеке Python есть модуль csv. Рассмотрим примеры работы с этим модулем.

```python
# Простое считывание строк

import csv

with open("students.csv", encoding="utf-8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=",")

    for index, row in enumerate(csv_reader):
        print(f"Строка {index}: {row}")
```

```python
# Простая запись строк

import csv

with open("students.csv", "a", encoding="utf-8") as csv_file:
    csv_writer = csv.writer(csv_file, delimiter=",")

    students_to_add = [
        ("Александр", "Александров", "642-2"),
        ("Владимир", "Владимиров", "642-2"),
    ]

    # Метод для записи набора строк
    csv_writer.writerows(students_to_add)

    # Метод для записи одной строки
    csv_writer.writerow(("Петр", "Петров", "345-2"))
```

```python
# Считывание словарей

import csv

with open("students.csv", encoding="utf-8") as csv_file:
    dict_reader = csv.DictReader(csv_file, delimiter=",")

    for index, row in enumerate(dict_reader):
        print(f"Строка {index}: {row}")
```

```python
# Запись словарей

import csv

with open("students.csv", "a", encoding="utf-8") as csv_file:
    csv_writer = csv.DictWriter(csv_file, delimiter=",", fieldnames=("Имя", "Фамилия", "Группа"))

    students_to_add = [
        {"Имя": "Александр", "Фамилия": "Александров", "Группа": "642-2"},
        {"Имя": "Владимир", "Фамилия": "Владимиров", "Группа": "642-2"},
    ]

    # Записываем заголовок таблицы
    csv_writer.writeheader()

    # Метод для записи набора словарей
    csv_writer.writerows(students_to_add)

    # Метод для записи одного словаря
    csv_writer.writerow({"Имя": "Петр", "Фамилия": "Петров", "Группа": "345-2"})
```

### 14.2.2 logging

***Логирование*** (журналирование) — это процесс автоматической записи событий, действий пользователей и ошибок, происходящих в работе ПО или IT-систем, в специальные файлы (логи).

Для организации системы логирования Python предоставляет стандартный модуль logging.

```python
import logging

logging.basicConfig(level=logging.DEBUG)
logging.debug("Строка для вывода в лог")
```

Модуль logging предлагает следующее разделение лог-сообщений по уровням "важности":
<ul>
<li>"DEBUG" - числовое представление: 10 - сообщение содержит полезную информацию для отладки</li>
<li>"INFO" - числовое представление: 20 - сообщение содержит базовую\дежурную информацию о том, что происходит в программе</li>
<li>"WARNING" - числовое представление: 30 - сообщение призывает администратора\разработчика обратить внимание на работу программы</li>
<li>"ERROR" - числовое представление: 40 - сообщение содержит сведения об ошибке</li>
<li>"CRITICAL" - числовое представление: 50 - сообщение содержит сведения о критической ошибке, которая может привести к полной остановке системы</li>
</ul>

```python
logging.debug("Отправляем сообщение для дебага")
logging.info("Отправляем сообщение о нормальном функционировании системы")
logging.warning("Отправляем сообщение о возникшей аномалии")
logging.error("Отправляем сообщение о возникшей ошибке")
logging.critical("Отправляем сообщение о возникшей критической ошибке")
```

Модуль logging позволяет задавать формат лог-сообщения. Для этого определен набор ключевых слов, которые нужно использовать в особой строке, передаваемой в качестве конфигурации. Полный перечень можно увидеть в документации: https://docs.python.org/3/library/logging.html#logrecord-attributes

В примере воспользуемся только некоторыми из определенных ключевых слов.

```python
# В строке формата можно использовать разные варианты указания параметров (%, $, {})

# Строка форматирования с %
logging.basicConfig(format="%(levelname)s:%(name)s:%(message)s")
# Строка форматирования с {}. Дополнительно задаем параметр style
logging.basicConfig(format="{levelname}:{name}:{message}", style="{")

logging.warning("Предупреждение!")
```

```python
# Метод .basicConfig может не сработать в этом блоке Jupyter Notebook
# Это связано с особенностями работы интерпретатора в режиме Jupyter Notebook, а так же с тем, что basicConfig может быть задан только единожды

logging.basicConfig(
    format="{asctime} - {levelname} - {message}",
    style="{",
    datefmt="%Y-%m-%d %H:%M",
)

logging.warning("Предупреждение с указанием времени вывода.")
```

В примерах мы использовали базовый (root) Logger для вывода сообщений, который объявлен прямо в модуле logging. Его настройка и вызов доступны через методы .basicConfig, logging.warning, logging.info и т.д.. Для учебной демонстрации этого достаточно, но на практике мы стремимся задать свой объект Logger-а для каждой части нашей системы.

Разберем как это сделать (а, заодно, добъемся вывода времени формирования лог-сообщения без необходимости перезапускать сессию интерпретатора):

```python
# Создаваемому Logger-у можно присвоить любое имя. Достаточно передать строчку с именем в качестве аргумента getLogger
# В качестве правила хорошего тона рекомендуется использовать имя модуля, получаемое из атрибута __name__
my_logger = logging.getLogger(__name__)
my_logger.warning("Это мой Logger!")
```

```python
# Logger при создании получает настройки уровня срабатывания от родительскго Logger-а
# Если не указано иное, родительским Logger-ом будет root (по умолчанию уровень для root это 30\WARNING, но выше мы меняли basicConfig на уровень 10\DEBUG)

# Проверяем уровень срабатывания (ожидаем 10\DEBUG или 30\WARNING в зависимости от выполнения ячейки с кодом вызова .basicConfig(level=logging.DEBUG))
print("Ближайший ненулевой уровень для my_logger: ", my_logger.getEffectiveLevel())

# Проверка уровня срабатывания через атрибут .level покажет "неожиданное" значение 0 (особый уровень NOTSET)
# Отличие связано с тем, что .level отображает заданный для данного конкретного объекта уровень срабатывания
# А .getEffectiveLevel показывает либо уровень .level, если он не равен 0, либо ближайший ненулевой уровень в цепочке родителей Logger-а
print("Заданный уровень для my_logger: ", my_logger.level)
print()

# Зададим уровень для объекта
my_logger.setLevel(logging.INFO)
print("Ближайший ненулевой уровень для my_logger: ", my_logger.getEffectiveLevel())
print("Заданный уровень для my_logger: ", my_logger.level)
```

Модуль logging предлагает гибкую схему для настройки наших Logger-ов. Она заключается в идее подключения к Logger-у отдельных объектов класса Handler. Каждый Handler может сохранять свои настройки формата вывода, места вывода и даже уровня срабатывания. 

Если к одному Logger-у подключено несколько Handler-ов, то единожды вызвав команду отправки лог-сообщения мы сможем отправить это лог-сообщение в разных форматах в разные места, т.е. нам не придётся в коде прописывать несколько последовательных вызовов вида:
<pre>
.warning("Сообщение", в_консоль)
.warning("Сообщение", в_файл)
.warning("Сообщение", в_другой_файл)
.warning("Сообщение", stderr)
</pre>
Это очень удобно. Рассмотрим как это работает на практике:

```python
# Создаем файловый Handler, параметры filename, mode, encoding соответствуют параметрам функции open
info_file_handler = logging.FileHandler("program.log", mode="a", encoding="utf-8")

err_file_handler = logging.FileHandler("errors.log", mode="a", encoding="utf-8")
# err_file_handler будет срабатывать только на сообщения уровня ERROR и выше, т.е. CRITICAL
err_file_handler.setLevel(logging.ERROR)

# Создаем Handler потока, по умолчанию StreamHandler использует системный поток stderr
stderr_handler = logging.StreamHandler()

my_logger.addHandler(info_file_handler)
my_logger.addHandler(err_file_handler)
my_logger.addHandler(stderr_handler)

# Теперь передача сообщения через my_logger будет перенаправлена в несколько Handler-ов
my_logger.info(msg="Подключили Handler. Ошибок не обнаружено.")
```

Если мы захотим выполнить форматирование вывода для Logger, мы снова не сможем воспользоваться методом .basicConfig, так как для настройки формата необходимо использовать объект класса Formatter.

Formatter подключается к объектам Handler, что позволяет задавать разные форматы для разных направлений вывода лог-сообщений.

```python
import logging
# Создаем форматтер
formatter = logging.Formatter(
    "{asctime} - {levelname} - {message}",
    style="{",
    datefmt="%Y-%m-%d %H:%M",
)

stderr_handler.setFormatter(formatter)

my_logger.info("Теперь вывод в Handler stderr_handler будет содержать время и уровень сообщения.")
```

Основная фильтрация лог-сообщений выполняется по уровню Logger-а и каждого отдельного Handler-а. Но, если нам нужно усложнить логику срабатывания фильтра, мы можем это сделать с помощью подключения своего фильтра в методе .addFilter.

Фильтр можно создать с помощью создания класса (класс потомок logging.Filter, либо класс с методом .filter) или отдельной функции, которая возвращает bool.

```python
from logging import LogRecord

# Фильтр пропускающий только сообщения уровня INFO
def custom_filter(record: LogRecord) -> bool:
    return record.levelname == "INFO"

stderr_handler.addFilter(custom_filter)

my_logger.debug("Это сообщение не попадет в поток stderr")
my_logger.info("Это сообщение попадет в поток stderr")
```

### 14.2.3 pickle

Иногда возникает необходимость сохранить текущее состояние программы и либо записать его в файл, либо отправить по сети. Процесс преобразования значений, сохраненных в оперативной памяти в поток байт\строк для записи на диск или передачи по сети называется **сериализацией**. Соответственно, обратный процесс называется **десериализацией**.

Python предлагает несколько стандартных модулей для выполнения сериализации\десериализации:
<ul>
<li>marshal</li>
<li>json</li>
<li>pickle</li>
</ul>

Модуль <code>marshal</code> используется самим интерпретатором и необходим для обработки файлов .pyc. Его использование не рекомендуется (в том числе из-за официального преупреждения в тексте документации о том, что формат работы marshal может быть изменен без возможности обратной совместимости).

Модуль <code>json</code> уже был рассмотрен вместе с темой "Модули".

Поэтому здесь кратко рассмотрим модуль <code>pickle</code>.

Модуль pickle позволяет выполнить сериализацию значений объектов в бинарный формат. 

Если наши значения должны быть читаемыми и десериализуемыми в других языках программирования, следует использовать модуль json. Во всех остальных случаях лучше воспользоваться pickle, так как он может выполнить сериализацию большего количества объектов разных классов (в том числе самописные классы).

Однако, если необходимо выполнить сериализацию функций, генераторов и lambda объектов, следует воспользоваться сторонней библиотекой <code>dill</code>. Так как pickle не умеет выполнять сериализацию подобных объектов.

```python
from dataclasses import dataclass
import pickle

@dataclass
class User:
    login: str
    name: str
    surname: str
    region: str

user_1 = User("ivan_ivan", "Иван", "Иванов", "Москва")

# Метод .dumps возвращает bytes после выполнения сериализации
pickled_user = pickle.dumps(user_1)
print(f"Объект user_1 после сериализации: {pickled_user}\n")

# Метод .loads ожидает bytes и выполняет десериализацию
unpickled_user = pickle.loads(pickled_user)
print(f"Объект после десериализации: {unpickled_user}")
```

```python
with open("pickled_user", "wb+") as out:
    # Метод .dump записывает результат сериализации в файл
    pickle.dump(user_1, out)

    out.seek(0)

    # Метод .load ожидает файл, а не bytes
    unpickled_user = pickle.load(out)
    print(f"Объект после десериализации из файла: {unpickled_user}")
```

```python
class SomeClass:
    def __init__(self):
        pass

    # Вызов .dump\-s приводит к вызову метода __getstate__
    # При желании мы можем настроить то, что будет передано в pickle при сериализации
    # Например, мы можем удалить из объекта атрибут, содержащий несериализуемые значения
    def __getstate__(self):
        return self.__dict__

    # Вызов .load\-s приводит к вызову метода __setstate__
    # Это приводит к тому, что выполнять десериализацию непроверенных объектов может быть опасно
    def __setstate__(self, state):
        print("Выполняю вредоносный код в этом методе...")


strange_object = SomeClass()
pickled_object = pickle.dumps(strange_object)
```

```python
# Выполнять десериализацю объектов из неизвестного источника - крайне опасно!

unpickled_object = pickle.loads(pickled_object)
```

### 14.2.4 json, collections, itertools (повторение материала из темы "Модули")

Примеры использования модулей json, collections и itertools уже были приведены в теме "Модули". В этот Jupyter Notebook они были скопированы для собрания всего материала в одном "месте".

```python
# json - работа с JSON
import json

# Python объект -> JSON строка
data = {
    "name": "Иван",
    "age": 25,
    "courses": ["Python", "SQL", "Git"],
    "is_student": True
}

json_string = json.dumps(data, ensure_ascii=False, indent=2)
print("Python -> JSON:")
print(json_string)

# JSON строка -> Python объект
parsed = json.loads(json_string)
print(f"\nJSON -> Python: {parsed['name']}, возраст {parsed['age']}")
```

```python
# collections - продвинутые структуры данных
from collections import Counter, defaultdict, namedtuple

# Counter - подсчет элементов
text = "абракадабра"
letter_count = Counter(text)
print(f"Подсчет букв: {letter_count}")
print(f"Самые частые: {letter_count.most_common(3)}")

# defaultdict - словарь со значением по умолчанию
word_lengths = defaultdict(list)
words = ["кот", "дом", "слон", "мир", "код", "муха"]
for word in words:
    word_lengths[len(word)].append(word)

print(f"\nСлова по длине: {dict(word_lengths)}")

# namedtuple - именованный кортеж
Point = namedtuple('Point', ['x', 'y'])
p = Point(10, 20)
print(f"\nТочка: x={p.x}, y={p.y}")
```

```python
# itertools - инструменты для итераций
from itertools import combinations, permutations, cycle, islice

# Комбинации
items = ['A', 'B', 'C']
print(f"Комбинации по 2: {list(combinations(items, 2))}")

# Перестановки
print(f"Перестановки по 2: {list(permutations(items, 2))}")

# Бесконечный цикл (берем первые 7 элементов)
colors = cycle(['красный', 'желтый', 'зеленый'])
print(f"Светофор: {list(islice(colors, 7))}")
```

4. Группирует товары в словарь вида {категория:список товаров категории}.
5. Сохраняет итоговый словарь в бинарный файл.
6. При следующем запуске восстанавливает объект из бинарного файла и выполняет дальнейшую работу с учетом восстановленного словаря.
